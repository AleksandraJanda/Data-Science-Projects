```python
corr_columns = correlations.columns
corr_columns
```

```python
# seek for columns where correlations are below -0.2 and above 0.2 (weak negative / positive correlations)
selected_columns = []
for col in corr_columns:
    column_data = correlations[col].sort_values()
    min = column_data.iloc[0]
    max = column_data.iloc[-2]
    if min < -0.2 or max > 0.2:
        selected_columns.append(col)
selected_columns
```

<h3>Model</h3>

```python
# seek for columns where correlations are irrelevant - to drop
irrelevant_columns = []
for col in corr_columns:
    column_data = correlations[col].sort_values()
    min = column_data.iloc[0]
    max = column_data.iloc[-2]
    if min >= -0.2 and max <= 0.2:
        irrelevant_columns.append(col)
irrelevant_columns
```

```python
model = df.copy()
```

```python
model.drop(columns=['job_role',
                    'company_size',
                    'meetings_per_day',
                    'deadlines_missed',
                    'screen_time_hours',
                    'caffeine_intake', 
                    'burnout_level'], inplace=True)
```

```python
genders = pd.get_dummies(model['gender']).astype(int)
genders.head()
```

```python
modes = pd.get_dummies(model['work_mode']).astype(int)
modes.head()
```

```python
model = pd.concat([model,genders,modes], axis=1)
```

```python
model.drop(columns=['gender','work_mode'], inplace=True)
```
